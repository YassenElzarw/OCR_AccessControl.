```python
reference_path = '/content/drive/MyDrive/Reference.JPG'
```

```python
reference=Image.open(reference_path)
```

```python
reference_text=pytesseract.image_to_string(reference)
```

```python
reference_text=reference_text.strip()
```

```python
uploaded = files.upload()
```

```python
file_name=""
```

```python
for i in uploaded.keys():
  file_name+=i
```

```python
input_image=Image.open(file_name)
```

```python
gray_password_input= input_image.convert('L')
```

```python
password_text=pytesseract.image_to_string(gray_password_input)
```

```python
password_text=password_text.strip()
```

```python
display(input_image,gray_password_input)
```

```python
nlp = spacy.load("en_core_web_sm")
doc1 = nlp(reference_text)
doc2 = nlp(password_text)
similarity_score=doc1.similarity(doc2)
```

```python
if password_text==reference_text:
  print("Access Granted")
else:
  print("Access Denied")

print("Similarity Score = ",similarity_score*100,"%")
```
